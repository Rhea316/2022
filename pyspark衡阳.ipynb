```python
import time
print(time.strftime('%Y-%m-%d %H:%M:%S'))
from pyspark.sql import SparkSession
def dataAnalyse(file_path):
    spark = SparkSession.builder.appName('pyspark').getOrCreate()
    csvfile = file_path
    '''
    #inferSchema——Automatically infer data types自动识别数据类型
    #timestampFormat——加载数据之前指定时间戳格式
    '''
    df = spark.read.option('inferSchema','true')\
    .option("timestampFormat", "yyyy-MM-dd hh:mm:ss")\
    .option('header','true').option('encoding','gbk')\
    .csv(csvfile)
    print('read ',time.strftime('%Y-%m-%d %H:%M:%S'))
    return df
df_2022 = dataAnalyse('G:\\衡阳案件数据v2\\2022-至今\\2022至今.csv')
df_2021 = dataAnalyse('G:\\衡阳案件数据v2\\2022-至今\\2021-2022.csv')
df_2020 = dataAnalyse('G:\\衡阳案件数据v2\\2022-至今\\2020-2021.csv')
df_2019 = dataAnalyse('G:\\衡阳案件数据v2\\2022-至今\\2019-2020.csv')
df_2018 = dataAnalyse('G:\\衡阳案件数据v2\\2022-至今\\2018-2019.csv')
df_总= df_2018.unionByName(df_2019).unionByName(df_2020).unionByName(df_2021).unionByName(df_2022)
print('汇总Done! ',time.strftime('%Y-%m-%d %H:%M:%S'))
# print(df_总.count())
from pyspark.sql.types import TimestampType, IntegerType, DoubleType, StringType, StructType, StructField
df_总 = df_总.withColumn("日期时间", df_总["日期时间"].cast(StringType()))
print('StringType ',time.strftime('%Y-%m-%d %H:%M:%S'))
```

```python
import pandas as pd
import numpy as np
import re
import datetime
import os

def index向后N位的index(df_序列, j_一方, no_N):
    for k in df_序列.index:
        if k > j_一方:
            no_序列号 = len(df_序列.loc[:k])
            no_目标序列号 = no_序列号 + no_N
            if no_目标序列号 >= len(df_序列):
                index_目标 = df_序列.index.max()
            else:
                index_目标 = df_序列.reset_index().iloc[no_目标序列号 - 1]['index']
            break
    return index_目标

def 一方匹配另一方(df_一方, all_另一方, tablename_索引, sheetname_层数):
    df_一方.index = df_一方['序号'].to_list()
    df_一方 = df_一方[df_一方['借贷标志'] == '进'].drop_duplicates(subset=['序号'], keep='last')
    df_匹配 = pd.DataFrame()
    df_未匹配 = pd.DataFrame()
    df_另一方 = pd.DataFrame(all_另一方[(all_另一方['借贷标志'] == '出')])
    print(str(len(df_一方)) + "条，最大索引为" + str(df_一方.index.max()), end=' ')

    for j in df_一方.index:
        for N in range(100):
            x = index向后N位的index(df_另一方, j, N)

            df_一进一出 = df_另一方[(df_另一方['序号'] <= x) & (df_另一方['序号'] > j) \
                             & (df_另一方['查询卡号'] == df_一方['查询卡号'].loc[j]) \
                             & (df_另一方['查询卡号'] != df_另一方['交易对方卡号']) \
                             ]
            if df_一进一出['交易金额'].sum() > df_一方['交易金额'].loc[j] * (1 - 0.10) :  ############下浮10%###########
                break
            else:
                continue
        if len(df_一进一出) > 0:
            merge_一进一出 = pd.merge(df_一方.loc[[j], :], df_一进一出, how='outer')
            df_匹配 = pd.concat([df_匹配, merge_一进一出])
            print(str(j) + "-" + str(x), end=' ')
        elif len(df_一进一出) == 0:
            df_未匹配 = pd.concat([df_未匹配, df_一方.loc[[j], :] ])
            print(str(j) + '失败', end=' ')
    if not os.path.isfile(output_path + tablename_索引 + "追踪转移.xlsx"):
        df_匹配.to_excel(output_path + tablename_索引 + "追踪转移.xlsx", sheet_name=sheetname_层数, index=False)
    else:
        with pd.ExcelWriter(output_path + tablename_索引 + "追踪转移.xlsx", mode='a',if_sheet_exists = 'new', engine='openpyxl') as writer:
            df_匹配.to_excel(writer, sheet_name=sheetname_层数, index=False)
    return df_匹配

def AB匹配(df_A, df_B, tablename_索引, sheetname_层数):
    df_A.index = df_A['序号'].to_list()
    df_A出B进 = pd.DataFrame()
    df_A出B空 = pd.DataFrame()
    df_A出 = df_A[df_A['借贷标志'] == '出'].drop_duplicates(subset=['序号'], keep='last')
    for k in df_A出.index:
        df_B进 = df_B[(df_B['借贷标志'] == '进') \
                     & (df_B['交易对方卡号'].isin(df_A出['查询卡号'])) \
                     & (df_B['交易对方卡号']!= df_B['查询卡号']) \
                     & (df_B['交易时间'] <= str(
            datetime.datetime.strptime(df_A出['交易时间'].loc[k], '%Y-%m-%d %H:%M:%S') + datetime.timedelta(seconds=60))) \
                     & (df_B['交易时间'] >= str(
            datetime.datetime.strptime(df_A出['交易时间'].loc[k], '%Y-%m-%d %H:%M:%S') + datetime.timedelta(seconds=-1))) \
                     & (df_B['交易金额'] == df_A出['交易金额'].loc[k]) \
                     ]
        if len(df_B进) > 0:
            merge_A出B进 = pd.merge(df_A出.loc[[k], :], df_B进, how='outer')
            df_A出B进 = pd.concat([df_A出B进, merge_A出B进])
            print(k, end=' ')
        elif len(df_B进) == 0 and df_B[df_B['查询卡号'].isin(df_A出.loc[[k], :]['交易对方卡号'])].shape[0]>0:
            df_B进 = df_B[(df_B['借贷标志'] == '进') \
                         & (df_B['交易对方卡号'].isin(df_A出['查询卡号'])) \
                         & (df_B['交易时间'] <= str(
                datetime.datetime.strptime(df_A出['交易时间'].loc[k][0:10]+' 23:59:59', "%Y-%m-%d %H:%M:%S"))) \
                         & (df_B['交易时间'] >= str(
                datetime.datetime.strptime(df_A出['交易时间'].loc[k][0:10], '%Y-%m-%d'))) \
                         & (df_B['交易金额'] == df_A出['交易金额'].loc[k]) \
                         ]
            if len(df_B进) > 0:
                time_A_struct = datetime.datetime.strptime(df_A出.loc[k,'交易时间'], "%Y-%m-%d %H:%M:%S")
                min_time_diff = df_B进['交易时间'].map(
                    lambda x:abs(datetime.datetime.strptime(x, "%Y-%m-%d %H:%M:%S") - time_A_struct).total_seconds()).min()
                df_B最优进 = df_B进[df_B进['交易时间'] == str(time_A_struct + datetime.timedelta(seconds=min_time_diff))]
                if len(df_B最优进)==0:
                    df_B最优进 = df_B进[df_B进['交易时间'] == str(time_A_struct - datetime.timedelta(seconds=min_time_diff))]
                merge_A出B进 = pd.merge(df_A出.loc[[k], :], df_B最优进, how='outer')
                df_A出B进 = pd.concat([df_A出B进,merge_A出B进])
                print(str(k),'最优时间差',min_time_diff,'秒', end=' ')
            elif len(df_B进) == 0:
                df_A出B空 = pd.concat([df_A出B空,df_A出.loc[[k], :]])
                print(str(k) + '对方'+str(df_B[df_B['查询卡号'].isin(df_A出.loc[[k], :]['交易对方卡号'])].shape[0])+'条中匹配失败', end=' ')
        else:
            print(str(k) + '对方调单'+str(df_B[df_B['查询卡号'].isin(df_A出.loc[[k], :]['交易对方卡号'])].shape[0])+'条', end=' ')
    return df_A出B进


def digui多层(df_层出层进, tablename_索引):
    if len(df_层出层进) > 0:
        df_某层 = 一方匹配另一方(df_层出层进, all_流水完整, tablename_索引, '层')
        if len(df_某层) > 0:
            df_某层A出B进 = AB匹配(df_某层, all_流水完整, tablename_索引, '层')
            digui多层(df_某层A出B进, tablename_索引)

def 合并多sheet24(多sheet_filename, tablename_索引):
    all_流水 = pd.DataFrame()
    Efile = pd.ExcelFile(多sheet_filename)
    sheet_list = pd.ExcelFile(多sheet_filename).sheet_names
    for j in range(0, len(sheet_list)):
        if sheet_list[j] != '':
            df = Efile.parse(sheet_name=sheet_list[j], dtype='object')
            df['深度'] = sheet_list[j]
            all_流水 = pd.concat([all_流水,df])
            print(str('') + '-' + str(j))
        else:
            continue
    all_流水.to_excel(output_path + tablename_索引 + "追踪转移合并.xlsx", index=False)
    return all_流水
```
