```python
def preprocess(series):

    entity_to_id = {}
    id_to_entity = {}
    for entity in series:
        if entity not in entity_to_id:
            new_id = len(entity_to_id)
            entity_to_id[entity] = new_id
            id_to_entity[new_id] = entity

    corpus = np.array([entity_to_id[w] for w in series])

    return corpus, entity_to_id, id_to_entity

corpus, entity_to_id, id_to_entity = preprocess(df_embs[0])
print(corpus)
print(entity_to_id)
print(id_to_entity)
```

```python
def most_similar(query, entity_to_id, id_to_entity, C, top=5):
    '''相似单词的查找

    :param query: 查询词
    :param word_to_id: 从单词到单词ID的字典
    :param id_to_word: 从单词ID到单词的字典
    :param word_matrix: 汇总了单词向量的矩阵，假定保存了与各行对应的单词向量
    :param top: 显示到前几位
    '''
    if query not in entity_to_id:
        print('%s is not found' % query)
        return

    print('\n[query] ' + query)
    query_id = entity_to_id[query]
    query_vec = C[query_id]

    vocab_size = len(id_to_entity) #7

    similarity = np.zeros(vocab_size) #array([0., 0., 0., 0., 0., 0., 0.])
    for i in range(vocab_size):
        similarity[i] = cosine(C[i], query_vec)
    out = {}
    count = 0
    for i in (-1 * similarity).argsort():
        if id_to_entity[i] == query:
            continue
        print(' %s: %s' % (id_to_entity[i], similarity[i]))
        out[count] = ' %s: %s' % (id_to_entity[i], similarity[i])  #输出到字典out
        count += 1
        if count >= top:
            return out

most_similar('25', entity_to_id, id_to_entity, C, top=5)
```
