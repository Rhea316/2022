```python
import pprint

import pymongo
myclient=pymongo.MongoClient("mongodb://localhost:27017/")
dblist=myclient.list_database_names()
drug_bp=myclient["painso"]
test_examine=drug_bp["day_air"]

test_find_all=test_examine.find()

structure ={}
for test_find_one in test_find_all:
    # 提炼一个表的表结构
    keys = test_find_one.keys()
    keys = list(keys)
    for key in keys:
        if key not in structure:
            structure[key]={}
        if isinstance(test_find_one[key], dict):
            keys1=list(test_find_one[key])
            for key1 in keys1:
                if key1 not in structure[key]:
                    structure[key][key1]={}
                if isinstance(test_find_one[key][key1], dict):
                    keys2=list(test_find_one[key][key1])
                    for key2 in keys2:
                        if key2 not in structure[key][key1]:
                            structure[key][key1][key2]={}
                        if isinstance(test_find_one[key][key1][key2], dict):
                            keys3 = list(test_find_one[key][key1][key2])
                            for key3 in keys3:
                                if key3 not in structure[key][key1][key2]:
                                    structure[key][key1][key2][key3]={}
pprint.pprint(structure)
```

```python
l_record_path=[]
l_meta=[]
def flatten2(data_dict):

    for key,value in data_dict.items():
        if isinstance(value, dict):
            # 第二层要继续判断
            for j in data_dict[key].keys():
                if isinstance(data_dict[key][j], dict):
                    print('dict',key,j)
                    l=[]
                    l.append(key)
                    l.append(j)
                    l_meta.append(l)
                elif isinstance(data_dict[key][j], list):
                    print('list',key,j)
                    l_record_path.append(key)
                    l_record_path.append(j)
                else:
                    l2=[]
                    l2.append(key)
                    l2.append(j)
                    l_meta.append(l2)
        elif isinstance(value, list):
            l_record_path.append(key)
        else:
            l_meta.append(key)
    print(l_record_path,l_meta)

flatten2(data_dict)
print(l_meta)
```

```python
tmp=df_拆分.reset_index().set_index(['index']).stack()
tmp2=tmp.rename_axis(index=['index','elements'])
tmp2.name='value'
tmp2.reset_index()
```

```python
import re
import numpy as np
df_2 = pd.merge(df.reset_index(),tmp2.reset_index(),left_on = 'index',right_on = 'index', how = 'right')
df_2['values'] = df_2['values'].map(lambda x:len(x.split('}'))-1)
# pattern = r'"pollutant_code" : "(\w+)"'
# df_2['pollutant_code'] = df_2['value'].map(lambda  x:re.findall(pattern, x)[0])

pattern_1 = r'"pollutant_code" : "?(.+?)"'
df_2['pollutant_code'] = df_2['value'].map(lambda x:re.findall(pattern_1, x)[0])
pattern_2 = r'"avg_value" : "(-?\d+\.\d+)"'
df_2['avg_value'] = df_2['value'].map(lambda x:re.findall(pattern_2, x)).map(lambda x:x[0] if len(x)>0 else np.nan)
pattern_3 = r'"cou_value" : "(-?\d+\.\d+)"'
df_2['cou_value'] = df_2['value'].map(lambda x:re.findall(pattern_3, x)).map(lambda x:x[0] if len(x)>0 else np.nan)

df_2
```
