```python
import pandas as pd
df = pd.read_excel(r'C:\Users\zqy\Desktop\来源去向.xlsx',sheet_name = 'Sheet2')
import re
df = df.applymap(lambda x:re.sub('nan','',str(x)))
bd = pd.melt(df, id_vars = df.columns.to_list()[0:10],value_vars = df.columns.to_list()[10:], var_name='type',value_name='value')
bd_1 = bd['type'].str.split('&',expand = True)
df_1 = pd.merge(bd,bd_1,how = 'left',left_index = True ,right_index = True)
df_1.drop(df_1[df_1['value']==''].index,axis = 0).to_excel(r'C:\Users\zqy\Desktop\转置完来源去向.xlsx',index = None)
```

```python
df_店 = pd.read_excel(r'C:\Users\zqy\Desktop\转置完.xlsx',dtype = object)
df_店 = df_店.applymap(lambda x:re.sub('nan','',str(x)))
df_店['对手方向'] = df_店['12月&时间&事项'].map(lambda x : '出' if x == '入' else '入')
df_店['左']= df_店['对手方向']+df_店['type']

df_店
```

```python
df_对手 = df_1.drop(df_1[df_1['value']==''].index,axis = 0)
df_对手['右']= df_对手['12月&时间&事项']+df_对手['type']
df_对手
```

```python
df_本方对手 = pd.merge(df_店,df_对手,how = 'right',left_on = '左' ,right_on = '右')
```

```python
df_本方对手.to_excel(r'C:\Users\zqy\Desktop\本方来源去向.xlsx',index = None)
df_本方对手 
```
