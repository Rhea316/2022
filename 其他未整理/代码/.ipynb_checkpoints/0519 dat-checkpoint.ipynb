```python
import os
str1 = '流水'
file_list = []
os.chdir(r'C:\Users\admin\Desktop\源csv')
for path_name, dir, files_name in os.walk(os.getcwd()):
    for file in files_name:
        if str1 in file:
            file_list.append(os.path.join(path_name,file))
print('\n'.join(file_list))

output = 'C:\\Users\\admin\\Desktop\\'

import pandas as pd

df_流水 = pd.DataFrame()
for i in range(len(file_list)):
    df = pd.read_csv(file_list[i],encoding = 'utf-8',dtype = object)
    df = df.applymap(lambda x:str(x))
    df_流水 = df_流水.append(df)
    print(i, end = ' ')
print('流水Done')

df_3号令_1 = pd.read_csv(r'C:\Users\admin\Desktop\3号令001.csv', encoding = 'utf-8',dtype = object)
df_3号令_2 = pd.read_csv(r'C:\Users\admin\Desktop\3号令002.csv',encoding = 'utf-8',dtype = object)
df_3号令 = df_3号令_1.append(df_3号令_2)
df_3号令 = df_3号令.applymap(lambda x:str(x))
print('读')       
                      
df_3号令抽取 = pd.DataFrame( columns = df_3号令.columns)
df_3号令中缺失 = df_流水[~df_流水['客户编号'].isin(df_3号令['客户编号'])].drop_duplicates(subset = '客户编号',keep = 'first')
df_3号令中缺失 = df_3号令中缺失[df_3号令中缺失['客户编号']!=' ']
df_3号令抽取['客户编号'] = df_3号令中缺失['客户编号']

df_3号令中线索 = df_3号令[df_3号令['客户编号'].isin(df_流水['客户编号'])]

df_3号令完整线索 = df_3号令中线索.append(df_3号令抽取)
print('抽')

df_3号令完整线索= df_3号令完整线索.applymap(lambda x :'|' +str(x)+'|')
df_3号令完整线索.iloc[:,0] = df_3号令完整线索.iloc[:,0].map(lambda x: x[1:])
df_3号令完整线索['最后一列'] = '|'
import re
df_3号令完整线索 = df_3号令完整线索.applymap(lambda x :re.sub('nan',' ',x))
df_3号令完整线索.to_csv(output +'3号令线索.txt',sep = '@',encoding = 'utf-8' ,header = None, index = None)

print('Done!')
```
