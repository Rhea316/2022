```python
output = 'C:\\Users\\admin\\Desktop\\excel\\流水拆分表\\'
# import pandas as pd
df = pd.read_csv(r'C:\Users\admin\Desktop\csv\P20210257_JS_FXQ_TRAN_20210511_0001_split50万-100万.csv',encoding = 'utf-8')
df = df.applymap(lambda x:str(x))
df
df_分编号 = df.groupby(['客户编号'],as_index = False).size()
for i in df_分编号.index:
    df_every = df.groupby(['客户编号'],as_index = False).get_group(df_分编号['客户编号'][i])
    df_every.to_excel(output +str(df_every['客户名称'].iloc[0])+str(df_every['客户编号'].iloc[0])+'.xlsx',index = None)
    print(i)
print('Done!')
```

```python
import os
str1 = 'TRAN'
file_list = []
os.chdir(r'C:\Users\admin\Desktop\csv')
for path_name, dir, files_name in os.walk(os.getcwd()):
    for file in files_name:
        if str1 in file:
            file_list.append(os.path.join(path_name,file))
print('\n'.join(file_list))

output = 'C:\\Users\\admin\\Desktop\\excel\\流水拆分表\\'

import pandas as pd

df_流水 = pd.DataFrame()
for i in range(len(file_list)):
    df = pd.read_csv(file_list[i],encoding = 'utf-8')
    df = df.applymap(lambda x:str(x))
    df_流水.append(df)
    print(i, end = ' ')

df_分编号 = df_流水.groupby(['客户编号'],as_index = False).size()
for i in df_分编号.index:
    df_every = df_流水.groupby(['客户编号'],as_index = False).get_group(df_分编号['客户编号'][i])
    df_every.to_excel(output +str(df_every['客户名称'].iloc[0])+str(df_every['客户编号'].iloc[0])+'.xlsx',index = None)
    print(i)
print('Done!')
```
