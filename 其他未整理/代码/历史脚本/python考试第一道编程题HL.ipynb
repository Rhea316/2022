```python
# -*- coding: utf-8 -*-
"""
Created on Fri Dec 14 09:45:40 2018

@author: jin
"""
import PyQt5
from PyQt5 import QtCore, QtGui, QtWidgets
from PyQt5.QtWidgets import *
from PyQt5.QtGui import *
from PyQt5.QtCore import *
import sys
import os
import xlrd
import csv
import pandas


class Merge_excelthread(QThread):
    trigger = pyqtSignal()

    def __int__(self):
        super(Merge_excelthread, self).__init__()

    def run(self):
        c = []
        failfile = []
        file_paths = []
        for k in range(listWidget.count()):
            file_paths.append(listWidget.item(k).text())
        countnumsuc = 0
        countsheetnumsuc = 0
        countsheetnumfail = 0

        for file_path in file_paths:
            try:
                workbook = xlrd.open_workbook(file_path)  # 打开工作表对象
                sheets = workbook.sheet_names()  # 获取excel中sheet页数量
                result = pandas.read_excel(file_path, dtype=str)

                for j in range(len(sheets)):  # 依次读取每个sheet页
                    a = []
                    worksheet = workbook.sheet_by_index(j)
                    for i in range(worksheet.nrows):
                        a.append(worksheet.row_values(i))
                    c.extend(a)
                    listWidget2.addItem(file_path+' sheet:'+sheets[j]+'合并完成')
                    countsheetnumsuc += 1
                    QApplication.processEvents()

                countnumsuc += 1

            except:
                listWidget2.addItem(file_path+' sheet:'+sheets[j]+'合并失败')
                failfile.append(file_path)
                countsheetnumfail += 1

        with open(savefilename, 'w', newline='') as f:
            csv_writer = csv.writer(f)
            csv_writer.writerows(c)
        failnum = len(list(set(failfile)))
        listWidget2.addItem('所有文件合并完毕: '+"合并成功%d个文件、%d个sheet" % (countnumsuc,
                                                                 countsheetnumsuc)+"合并失败%d个文件、%d个sheet" % (failnum, countsheetnumfail))

        self.trigger.emit()


class seperate_excel(QThread):
    trigger = pyqtSignal()

    def __int__(self):
        super(WorkThread, self).__init__()

    def run(self):

        file_paths = []
        for k in range(listWidget.count()):
            file_paths.append(listWidget.item(k).text())

        for file_path in file_paths:
            for i in range(len(file_path)):
                if file_path[-i] == '/':
                    FiLe_name = file_path[-(i-1):-5]
                    break

            try:
                workbook = xlrd.open_workbook(file_path)  # 打开工作表对象
                sheets = workbook.sheet_names()  # 获取excel中sheet页数量
                for j in range(len(sheets)):  # 依次读取每个sheet页
                    a = []
                    worksheet = workbook.sheet_by_index(j)
                    for i in range(worksheet.nrows):
                        a.append(worksheet.row_values(i))
                    savefilename1 = save_names1+'/'+FiLe_name+sheets[j]+'.csv'
                    with open(savefilename1, 'w', newline='') as f:
                        csv_writer = csv.writer(f)
                        csv_writer.writerows(a)

                    listWidget2.addItem(file_path+' sheet:'+sheets[j]+'拆分完成')
                    QApplication.processEvents()

            except:
                listWidget2.addItem(file_path+' sheet:'+sheets[j]+'拆分失败')

        self.trigger.emit()
        listWidget2.addItem('所有文件拆分完成')


class Merge_TxTthread(QThread):
    trigger = pyqtSignal()

    def __int__(self):
        super(Merge_TxTthread, self).__init__()

    def run(self):
        file_paths = []
        for k in range(listWidget.count()):
            file_paths.append(listWidget.item(k).text())
        with open(savefilename, 'w', newline='') as f2:
            for file_path in file_paths:
                try:
                    fk = open(file_path, 'r', encoding='utf-8')
                    f2.write(fk.read())
                    fk.close()
                except:
                    fk = open(file_path, 'r', encoding='gbk')
                    f2.write(fk.read())
                    fk.close()

                listWidget2.addItem(file_path+'合并完成')
                QApplication.processEvents()

        self.trigger.emit()
        listWidget2.addItem('所有文件合并完成')


class Drop_Duplicates(QThread):
    trigger = pyqtSignal()

    def __int__(self):
        super(Drop_Duplicates, self).__init__()

    def run(self):

        drop_filename = line_edit.text()
        save_drop_filename = line_edit2.text()
        if drop_filename[-4:] == ".xls" or drop_filename[-5:] == ".xlsx":
            f = pandas.read_excel(drop_filename, dtype=str)
            f.drop_duplicates(keep='first', inplace=True)

            if save_drop_filename[-4:] == ".xls" or save_drop_filename[-5:] == ".xlsx":
                f.to_excel(save_drop_filename, index=False)

            elif save_drop_filename[-4:] == ".txt" or save_drop_filename[-4:] == ".csv":
                f.to_csv(save_drop_filename, index=False)
        else:
            try:
                f = open(drop_filename, 'r', encoding='utf-8')
                c = f.readlines()
                d = []
                for cd in c:
                    if cd not in d:
                        d.append(cd)
                f2 = open(save_drop_filename, 'w')
                for st in d:
                    f2.write(st)
                f.close()
                f2.close()

            except:

                f = open(drop_filename, 'r', encoding='gbk')
                c = f.readlines()
                d = []
                for cd in c:
                    if cd not in d:
                        d.append(cd)
                f2 = open(save_drop_filename, 'w')
                for st in d:
                    f2.write(st)
                f.close()
                f2.close()

        self.trigger.emit()


def About_wrapper(func):
    def wrapper():
        if listWidget.count() == 0:
            QMessageBox.about(main_window, "提示", "当前未选择任何文件\n请先选择需要操作的文件")
        else:
            return func()
    return wrapper


def menu_signal(q):
    if (q.text() == "选择文件"):
        file_dialog = QFileDialog()
        file_dialog.setFileMode(QFileDialog.ExistingFiles)
        if file_dialog.exec_():
            filenames = file_dialog.selectedFiles()
            for filename in filenames:
                listWidget.addItem(filename)

    if (q.text() == "选择文件夹"):
        dirname = QFileDialog.getExistingDirectory()
        for parent, dirnames, filenames in os.walk(dirname):
            for filepath in filenames:
                file_path = parent + '/' + filepath
                listWidget.addItem(file_path)

    if (q.text() == "About"):
        version_message = '''
        本程序初步设想是将案件清洗时常用的功能整合起来。
        这样清洗数据时会更方便、免去每次启动python环境找到特定脚本清洗数据的麻烦
        同时也可以将小程序送给客户使用，便于客户在没有宝盒的时候提高工作效率。
        当前拥有功能：文件合并（excel\\txt\csv）、单文件去重（excel\\txt\csv）、excel表格sheet拆分。
        '''
        QMessageBox.about(main_window, "软件介绍", version_message)

    statusbar.showMessage('已选择'+str(listWidget.count())+'个文件')


def select_drop_file():
    file_dialog, filetype = QFileDialog.getOpenFileName()
    line_edit.setText(file_dialog)


def save_drop_file():
    file_dialog, filetype = QFileDialog.getSaveFileName()
    line_edit2.setText(file_dialog)


def drop_Duplicates():
    drop_filename = line_edit.text()
    save_drop_filename = line_edit2.text()

    if drop_filename == "":
        QMessageBox.about(main_window, "提示", "请选择需要去重的文件")
    elif save_drop_filename == "":
        QMessageBox.about(main_window, "提示", "请选择文件存放路径")

    else:
        dropThread.start()
        dropThread.trigger.connect(finish_work3)


@About_wrapper
def seperate_Excel():
    QMessageBox.about(main_window, "提示", "设置拆分后excel文件存放目录")
    global save_names1
    save_names1 = QFileDialog.getExistingDirectory()

    if save_names1 == '':
        return
    else:
        workThread1.start()
        workThread1.trigger.connect(finish_work2)


@About_wrapper
def Merge_TxT():
    QMessageBox.about(main_window, "提示", "设置合并后文件保存的位置")
    save_names, filetype = QFileDialog.getSaveFileName()
    if save_names == "":
        return
    else:
        global savefilename
        savefilename = save_names+'.csv'
        workThread2.start()
        workThread2.trigger.connect(finish_work)


def finish_work():
    QMessageBox.about(main_window, "提示", "合并完成")


def finish_work2():
    QMessageBox.about(main_window, "提示", "拆分完成")


def finish_work3():
    QMessageBox.about(main_window, "提示", "去重完成")


def Clear_select():
    listWidget.clear()


def Clear_finished():
    listWidget2.clear()


def btn_reEnable(i):
    if cb.currentText() == '':
        pushButton.setEnabled(False)
    else:
        pushButton.setEnabled(True)


def Mergeall():
    if cb.currentText() == '合并excel':
        Merge_excel()
    elif cb.currentText() == '合并txt/csv':
        Merge_TxT()
    else:
        return


@About_wrapper
def Merge_excel():
    QMessageBox.about(main_window, "提示", "设置合并后文件保存的位置")
    save_names, filetype = QFileDialog.getSaveFileName()
    if save_names == "":
        return
    else:
        global savefilename
        savefilename = save_names+'.csv'
        workThread.start()
        workThread.trigger.connect(finish_work)


def listwidget_rightMenu():
    rightMenu = QMenu(listWidget)
    # triggered 为右键菜单点击后的激活事件。这里slef.close调用的是系统自带的关闭事件。
    removeAction = QAction("移除", main_window, triggered=listwidget_delete)
    previewaction = QAction("预览", main_window, triggered=file_preview)

    rightMenu.addAction(removeAction)
    rightMenu.addAction(previewaction)
    rightMenu.exec_(QtGui.QCursor.pos())


def listwidget_delete():
    listWidget.takeItem(listWidget.currentRow())


def file_preview():
    QMessageBox.about(main_window, "提示", "功能未完善、敬请期待！")


if __name__ == "__main__":
    app = PyQt5.QtWidgets.QApplication(sys.argv)
    app.setWindowIcon(QIcon("C:\\Users\\admin\\Desktop\\exe\\death.ico"))
    main_window = PyQt5.QtWidgets.QMainWindow()
    main_window.resize(900, 600)
    main_window.setWindowTitle("hello world")
    main_window.setEnabled(True)

    listWidget = QtWidgets.QListWidget(main_window)
    listWidget.setGeometry(QtCore.QRect(30, 60, 300, 340))
    listWidget.setObjectName("listWidget")  # 放置一个列表框

    pushButton = QtWidgets.QPushButton(main_window)
    pushButton.setGeometry(QtCore.QRect(80, 440, 90, 28))
    pushButton.setText("合并所有文件")  # 放置一个按钮控件，并设置相关参数

    cb = QComboBox(main_window)
    cb.addItem('')
    cb.addItem('合并excel')
    cb.addItem('合并txt/csv')
    cb.setGeometry(QtCore.QRect(250, 440, 90, 28))
    cb.currentIndexChanged.connect(btn_reEnable)

    pushButton_2 = QtWidgets.QPushButton(main_window)
    pushButton_2.setGeometry(QtCore.QRect(390, 120, 111, 28))
    pushButton_2.setText("清空已选择")
    pushButton_3 = QtWidgets.QPushButton(main_window)
    pushButton_3.setGeometry(QtCore.QRect(390, 280, 111, 28))
    pushButton_3.setText("清空状态栏")

    pushButton_4 = QtWidgets.QPushButton(main_window)
    pushButton_4.setGeometry(QtCore.QRect(80, 500, 90, 28))
    pushButton_4.setText("拆分excel")

    pushButton_5 = QtWidgets.QPushButton(main_window)
    pushButton_5.setGeometry(QtCore.QRect(570, 510, 75, 23))
    pushButton_5.setText("执行")

    line_edit = QLineEdit(main_window)
    line_edit.setGeometry(QtCore.QRect(510, 440, 221, 20))
    line_edit.setPlaceholderText('需要去重的文件路径')

    line_edit2 = QLineEdit(main_window)
    line_edit2.setGeometry(QtCore.QRect(510, 480, 221, 20))
    line_edit2.setPlaceholderText('添加文件格式后缀(.txt)/(.xlsx)等')

    tool_button = QToolButton(main_window)
    tool_button.setGeometry(QtCore.QRect(770, 440, 40, 20))
    tool_button.setText("...")

    tool_button2 = QToolButton(main_window)
    tool_button2.setGeometry(QtCore.QRect(770, 480, 40, 20))
    tool_button2.setText("...")

    hline = QFrame(main_window)
    hline.setGeometry(QtCore.QRect(40, 400, 821, 20))
    hline.setFrameShape(QtWidgets.QFrame.HLine)
    hline.setFrameShadow(QtWidgets.QFrame.Sunken)

    hline2 = QFrame(main_window)
    hline2.setGeometry(QtCore.QRect(40, 540, 821, 20))
    hline2.setFrameShape(QtWidgets.QFrame.HLine)
    hline2.setFrameShadow(QtWidgets.QFrame.Sunken)

    vline = QFrame(main_window)
    vline.setGeometry(QtCore.QRect(30, 409, 20, 140))
    vline.setFrameShape(QtWidgets.QFrame.VLine)
    vline.setFrameShadow(QtWidgets.QFrame.Sunken)

    vline2 = QFrame(main_window)
    vline2.setGeometry(QtCore.QRect(370, 409, 20, 140))
    vline2.setFrameShape(QtWidgets.QFrame.VLine)
    vline2.setFrameShadow(QtWidgets.QFrame.Sunken)

    vline3 = QFrame(main_window)
    vline3.setGeometry(QtCore.QRect(850, 409, 20, 140))
    vline3.setFrameShape(QtWidgets.QFrame.VLine)
    vline3.setFrameShadow(QtWidgets.QFrame.Sunken)

    label = QtWidgets.QLabel(main_window)
    label.setGeometry(QtCore.QRect(30, 30, 161, 31))
    label.setText("已选择文件：")  # 放置标签设置参数

    label_2 = QtWidgets.QLabel(main_window)
    label_2.setGeometry(QtCore.QRect(560, 30, 161, 31))
    label_2.setText("工作状态：")

    label_3 = QtWidgets.QLabel(main_window)
    label_3.setGeometry(QtCore.QRect(250, 410, 90, 28))
    label_3.setText("选择文件类型：")

    label_4 = QtWidgets.QLabel(main_window)
    label_4.setGeometry(QtCore.QRect(570, 420, 61, 16))
    label_4.setText("<b>数据去重<\b>")

    label_5 = QtWidgets.QLabel(main_window)
    label_5.setGeometry(QtCore.QRect(410, 440, 91, 20))
    label_5.setText("选择文件：")

    label_6 = QtWidgets.QLabel(main_window)
    label_6.setGeometry(QtCore.QRect(410, 480, 91, 20))
    label_6.setText("去重后存放位置：")

    listWidget2 = QtWidgets.QListWidget(main_window)
    listWidget2.setGeometry(QtCore.QRect(560, 60, 300, 340))
    listWidget2.setObjectName("listWidget2")

    menubar = main_window.menuBar()  # menuBar()方法返回主窗口的QMenuBar对象
    File = menubar.addMenu('选择文件')
    File.addAction("选择文件")  # 添加菜单栏
    File.addAction("选择文件夹")

    Help = menubar.addMenu('Help')
    Help.addAction("About")

    #More = menubar.addMenu('其他功能')
    # More.addAction("文件去重")

    statusbar = QtWidgets.QStatusBar(main_window)
    statusbar.setObjectName("statusbar")
    main_window.setStatusBar(statusbar)

   # 定义功能区
    listWidget.setContextMenuPolicy(Qt.CustomContextMenu)
    listWidget.customContextMenuRequested[QtCore.QPoint].connect(
        listwidget_rightMenu)  # 设置listwidget的右键属性

    workThread = Merge_excelthread()
    workThread1 = seperate_excel()
    workThread2 = Merge_TxTthread()
    dropThread = Drop_Duplicates()
    File.triggered[QAction].connect(menu_signal)
    Help.triggered[QAction].connect(menu_signal)

    pushButton.setEnabled(False)
    pushButton.clicked.connect(Mergeall)
    pushButton_2.clicked.connect(Clear_select)
    pushButton_3.clicked.connect(Clear_finished)
    pushButton_4.clicked.connect(seperate_Excel)
    pushButton_5.clicked.connect(drop_Duplicates)
    tool_button.clicked.connect(select_drop_file)
    tool_button2.clicked.connect(save_drop_file)

    # 垂直布局类QVBoxLayout
    main_window.show()
    sys.exit(app.exec_())
```
