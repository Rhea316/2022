```python
import pandas as pd
import os
import time
import numpy as np

os.chdir(r"F:\新建文件夹")
path1 = os.getcwd()
str1 = "user_"
print(path1)
file_list = []

def search_file(path,str):
    for file in os.listdir(path):
        this_path = os.path.join(path,file)
        if os.path.isfile(this_path):
            if str in this_path:
                file_list.append(this_path)
        else:
            search_file(this_path,str)
    return file_list
print(search_file(path1,str1))
print(len(file_list))

def timestamp_to_date(time_stamp, format_string="%Y-%m-%d %H:%M:%S"):
    time_array = time.localtime(time_stamp)
    str_date = time.strftime(format_string, time_array)
    return str_date

all_data = pd.DataFrame()

for i in range(len(file_list)):
    print(i)
    l_data = pd.read_excel(file_list[i],encoding='utf-8',header=None)
    all_data = all_data.append(l_data)
    

df = pd.DataFrame(all_data)
df.columns =all_data.iloc[0]
df = df.drop([0])
df['addTime'] = df['addTime'].astype(int).apply(timestamp_to_date)
df['djtime'] = df['djtime'].fillna(0).apply(timestamp_to_date)
df['fuwstime'] = df['fuwstime'].fillna(0).apply(timestamp_to_date)
df['jdcstime'] = df['jdcstime'].fillna(0).apply(timestamp_to_date)
df['zhsytime'] = df['zhsytime'].fillna(0).apply(timestamp_to_date)
df['sqtime'] = df['sqtime'].fillna(0).apply(timestamp_to_date)
df['czsdtime'] = df['czsdtime'].fillna(0).apply(timestamp_to_date)
df['dongjtime'] = df['dongjtime'].fillna(0).apply(timestamp_to_date)

df.to_csv(r"F:\新建文件夹\合并.csv" ,encoding="utf-8",index=False)
print("Done!")
```
