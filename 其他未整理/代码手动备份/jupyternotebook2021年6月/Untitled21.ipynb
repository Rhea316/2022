所有Sheet批量读取合并df

```python
import pandas as pd
Efile =  pd.ExcelFile("E:\\泰州模型\\给维正.xls")
sheet_list = pd.ExcelFile("E:\\泰州模型\\给维正.xls").sheet_names
df = pd.DataFrame()
for i in range(0,len(sheet_list)):
    df_sheet = Efile.parse(sheet_name = sheet_list[i])
    df = df.append(df_sheet)
print(df)
```

标题、内容,去除前缀后缀\t

```python
df.columns = df.columns.str.strip()
for i in range(0,df.shape[1]):
    df.iloc[:,i] = df.iloc[:,i].astype(str).str.strip()
print(str(i+1)+'列去除前缀后缀')
print(df.head())
df.groupby('交易卡号')
```

给每个按交易卡号划分的组，index为组号

```python
df_S1 = df.groupby('交易卡号').size().reset_index()
df_S1
```

组号遍历循环，对每组，统计记录数Size、不重复记录数False、重复记录数True

```python
df_S2 = pd.DataFrame(columns = ['交易卡号','Size','False','True'])
for i in df_S1.index:
    df_every = df.groupby('交易卡号').get_group(df_S1['交易卡号'][i])
    bool_S = df_every.duplicated().value_counts()
    if len(bool_S)==1:
        df_S2.loc[i] = {'交易卡号':df_every['交易卡号'].iloc[0],'Size':df_every.shape[0],'False':bool_S[0],'True':0}
    elif len(bool_S)==2:
        df_S2.loc[i] = {'交易卡号':df_every['交易卡号'].iloc[0],'Size':df_every.shape[0],'False':bool_S[0],'True':bool_S[1]}
print(df_S2)
```

输出单卡每组去重统计表

```python
df_S2.to_excel("E:\\泰州模型\\单卡去重.xls",index = None)
print('Done!')
```
