```python
import pandas as pd
df = pd.read_csv("E:\\7.3徐州网赌\\流水户名.csv",encoding ="utf-8",engine = "python")
df.shape
```

```python
import numpy as np
df = df[df['交易时间']> '2020-01-01 00:00:00']
table_time = df.groupby(['交易卡号'],as_index = False).aggregate({'交易时间':np.min})
table_group = df.groupby(['交易卡号','收付标志'],as_index = False).aggregate({'交易金额':np.sum})
print(table_group)
print(type(table_group))
df_table = pd.merge(table_group[table_group['收付标志'] == '进'],table_group[table_group['收付标志'] == '出'],how = 'outer',on = '交易卡号')
print(df_table)
df_table['交易额总和'] = df_table['交易金额_x']+df_table['交易金额_y']
df_table['交易额净和'] = df_table['交易金额_x']-df_table['交易金额_y']
df_余额 = pd.read_excel("E:\\7.3徐州网赌\\单卡最后一笔交易余额.xlsx",encoding ="gbk")
df_analysis = pd.merge(df_余额,df_table,how = 'left',on = '交易卡号')
df_analysis = pd.merge(df_analysis,table_time,how = 'left',on = '交易卡号')
df_analysis.to_excel("E:\\7.3徐州网赌\\主账号列表统计分析2020以后.xlsx")
print('Done!')
```
