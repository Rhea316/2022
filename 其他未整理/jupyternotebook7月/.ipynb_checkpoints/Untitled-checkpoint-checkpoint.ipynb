```python
import os
str1 = ''
os.chdir(r"E:\7.3徐州网赌\20200626经侦来的数据原件总")
file_list = []
for path_name, dir, files_name in os.walk(os.getcwd()):
    for file in files_name:
        if str1 in file:
            file_list.append(os.path.join(path_name, file))
file_list
```

上面file_list中，xls、xlsx、csv格式文件读取

df_group表是path、columns两列统计表,all_data表是读取的所有文件合并

colunmns不同合并失败，故需要识别colunms再分组

```python
import pandas as pd
all_data = pd.DataFrame()
df_group = pd.DataFrame(columns = ['Path','Columns'])
type1 = '.xls'
type2 = '.csv'
for i in range(len(file_list)):
    print(i)
    if type1 in file_list[i]:
        df = pd.read_excel(file_list[i], encoding = "gbk")
        col_name = df.columns.tolist()
        df_group.loc[i] = {'Path':file_list[i],'Columns':tuple(col_name)}
        
        print(df.shape)
        print(df.duplicated().value_counts()) # 表内重复值行数 'True 数量'
        df['path'] = file_list[i]
        df['编号'] = df.index
        data_1 = df.drop_duplicates(subset = col_name,keep = 'last')
        all_data = all_data.append(data_1)
    elif type2 in file_list[i]:
        df = pd.read_csv(file_list[i], engine = "python")
        col_name = df.columns.tolist()      
        df_group.loc[i] = {'Path':file_list[i],'Columns':tuple(col_name)}
        
        print(df.shape)
        print(df.duplicated().value_counts()) # 表内重复值行数 'True 数量'
        df['path'] = file_list[i]
        df['编号'] = df.index
        data_1 = df.drop_duplicates(subset = col_name,keep = 'last')
        all_data = all_data.append(data_1)

print(all_data.shape)
```

grouped是按columns分组（数据结构特殊，不是DataFramen！）

grouped.size() 是series数据结构，index是组，值是每组计数

.get_group()返回的是每组明细，groupby是对Dataframe则返回的是Dataframe

按index[0] 、1、2…… 把文件分到不同组，得到不同分组好的文件路径group_0、group_1……

```python
# 示例，分组以后，指定第一组第201个文件

grouped = df_group.groupby('Columns') # 注意Column是col_name的元组数据结构值，groupby不支持列表结构的值
df_series = grouped.size()
print(len(df_series.index))
group_0 = grouped.get_group(df_series.index[0]) # 与第一组标题行columns格式一致的分一组
print(group_0.loc[200])
```

```python
import pandas as pd

df_group = pd.DataFrame(columns = ['Path','Columns','False','True'])
type1 = '.xls'
type2 = '.csv'
for i in range(len(file_list)):
#     print(i)
    if type1 in file_list[i]:
        df = pd.read_excel(file_list[i], encoding = "gbk")
        col_name = df.columns
        bool_S = df.duplicated().value_counts()
        if len(bool_S)==1:
            df_group.loc[i] = {'Path':file_list[i],'Columns':tuple(col_name),'False':bool_S[0],'True':0}
        elif len(bool_S)==2:
            df_group.loc[i] = {'Path':file_list[i],'Columns':tuple(col_name),'False':bool_S[0],'True':bool_S[1]}

    elif type2 in file_list[i]:
        df = pd.read_csv(file_list[i], engine = "python")
        col_name = df.columns
        bool_S = df.duplicated().value_counts()
        if len(bool_S)==1:
            df_group.loc[i] = {'Path':file_list[i],'Columns':tuple(col_name),'False':bool_S[0],'True':0}
        elif len(bool_S)==2:
            df_group.loc[i] = {'Path':file_list[i],'Columns':tuple(col_name),'False':bool_S[0],'True':bool_S[1]}

grouped = df_group.groupby('Columns')
print(grouped.size())
```

组0是银行交易流水，组5是账户开户信息

```python
groupname_list = list(grouped.size().index)
print(len(groupname_list))
group_0 = grouped.get_group(groupname_list[0])
group_0['文件序号I'] = group_0.index
group_0.to_excel("E:\\7.3徐州网赌\\流水文件目录.xlsx")
group_5 = grouped.get_group(groupname_list[5])
group_5['文件序号I'] = group_5.index
group_5.to_excel("E:\\7.3徐州网赌\\账户文件目录.xlsx")
print('Done!')
group_流水 = group_0
group_账户 = group_5
```

```python
all_data = pd.DataFrame()
all_account = pd.DataFrame()

for i in range(len(file_list)):
    if file_list[i] in list(group_流水['Path']):
        print(i)
        if type1 in file_list[i]:
            df = pd.read_excel(file_list[i], encoding = "gbk")
            col_name = df.columns.tolist()
            df['path'] = file_list[i]
            df['编号'] = df.index
            data_1 = df.drop_duplicates(subset = col_name,keep = 'last')
            all_data = all_data.append(data_1)
        elif type2 in file_list[i]:
            df = pd.read_csv(file_list[i], engine = "python")
            col_name = df.columns.tolist()
            df['path'] = file_list[i]
            df['编号'] = df.index
            data_1 = df.drop_duplicates(subset = col_name,keep = 'last')
            all_data = all_data.append(data_1)
    elif file_list[i] in list(group_账户['Path']):
        df_account = pd.read_csv(file_list[i], engine = "python")
        df_account['path'] = file_list[i]
        all_account = all_account.append(df_account)
print(all_data.shape)
print(all_account.shape)
```

```python

all_data.to_csv("E:\\7.3徐州网赌\\流水汇总.csv",index = None)
all_account.to_csv("E:\\7.3徐州网赌\\账户汇总.csv",index = None)
print('Done!')
```
