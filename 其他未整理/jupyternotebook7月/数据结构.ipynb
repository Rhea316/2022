```python
class Stack(object):             # 定义类
    def _init_(self,limit=10):
        self.stack = []          # []空列表
        self.limit = limit       # 属性self.limit 指向变量limit
    
    def push(self,data):
        if len(self.stack) >= self.limit:       #判断栈是否溢出
            raise IndexError('超出栈容量极限')  # raise手动引发的异常，基本语法raise [exceptionName [(reason)]]
        self.stack.append(data)   # 把变量data 进栈
        
    def pop(self):
        if self.stack:
            return self.stack.pop()  # 栈顶元素移出
        else:
            raise IndexError('pop from an empty stack') # raise手动引发的异常
    
    def peek(self):
        if self.stack:
            return self.stack[-1]   # 堆栈栈顶的元素
        
    def is_empty(self):
        return not bool(self.stack)  # 栈转换为布尔类型，not True 即False
    
    def size(self):
        return len(self.stack)   # 返回栈大小
```
