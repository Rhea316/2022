读取文件夹下全部文件

获得文件路径list

```python
import os
str1 = '交易明细信息'
os.chdir(r"E:\5.28职务侵占第二阶段\平台反馈")
file_list = []
for path_name, dir, files_name in os.walk(os.getcwd()):
    for file in files_name:
        if str1 in file:
            file_list.append(os.path.join(path_name, file))
file_list
```

读取文件xlsx、csv、txt

循环获得文件header、路径、行列数、重复行数

```python
import pandas as pd
type1 = 'xlsx'
type2 = 'csv'
type3 = 'txt'
for filename in file_list:
    if type1 in filename:
        df = pd.read_excel(filename, encoding = "gbk")
        print(df.columns)
        print(filename)
        print(df.shape)
        print(df.duplicated().value_counts())
    elif type2 in filename:
        df = pd.read_csv(filename, engine = "python")
        print(df.columns.str.strip())
        print(filename)
        print(df[df['交易卡号\t'].isin(['6227001280300129053'])])  #筛选交易卡号指定值
        print(df.shape)
        print(df.duplicated().value_counts())
    elif type3 in filename:
        df = pd.read_table(filename, encoding = "gbk")
        print(df.columns)
        print(filename)
        print(df.shape)
        print(df.duplicated().value_counts())
```
